```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
print('Halo')
```

```python
data = pd.read_csv('/content/train.csv')
```

```python
data.head(5)
```

```python
data.shape
```

```python
data.info()
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
data = data.dropna()
```

```python
data.isnull().sum()
```

```python
X = data.iloc[:,0]
X.head(5)
```

```python
y = data.iloc[:,-1]
y.head(5)
```

```python
plt.scatter(y,X)
plt.title('Salary vs price_range')
plt.xlabel('price_range')
plt.ylabel('Salary')
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.33,random_state=51)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
linear = LinearRegression()
linear.fit(np.array(X_train).reshape(-1,1),y_train)
```

```python
linear.coef_
```

```python
linear.intercept_
```

```python
Y_pred = linear.predict(np.array(X_test).reshape(-1,1))
```

```python
Y_pred
```

```python
score = r2_score(y_test,Y_pred)
score
```

```python
plt.scatter(X_test,y_test)
plt.plot(X_test,Y_pred,color='red')
plt.title('Salary vs price_range')
plt.xlabel('price_range')
plt.ylabel('Salary')
```
